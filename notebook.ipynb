## Data Preparation / Preprocessing

Kita akan melakukan binerisasi label target (`Status`) dan memastikan semua fitur sudah dalam bentuk numerik.

Tahapan yang dilakukan:
- Seleksi data (fokus pada label Dropout vs Graduate)
- Encoding fitur kategorikal
- Normalisasi/standarisasi fitur numerik
- Split data untuk training dan testing

### Seleksi data (fokus pada label Dropout vs Graduate)

```python
# Fokus hanya pada 2 kelas: Dropout dan Graduate
df = df[df['Status'].isin(['Dropout', 'Graduate'])]

# Label encoding target
df['Status'] = df['Status'].map({'Graduate': 0, 'Dropout': 1})
```

Encoding fitur kategorikal

```python
# Encode semua fitur kategorikal
df_encode = df.copy()
for col in df_encode.select_dtypes(include='object').columns:
    df_encode[col] = df_encode[col].astype('category').cat.codes
```

Memisahkan Fitur (X) dan Label atau target (Y)

```python
# Pisahkan fitur dan label
X = df_encode.drop('Status', axis=1)
y = df_encode['Status']
```

Melakukan Standarisasi terhadap fitur numerik

```python
# Standarisasi fitur numerik
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

Melakukan pembagian data untuk train dan test

```python
# Split data: train 80% - test 20%
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, stratify=y, random_state=42
)
```

## Modeling

Algoritma yang digunakan adalah **Random Forest Classifier**, yaitu algoritma ensemble berbasis pohon keputusan yang memiliki kemampuan yang baik untuk melakukan klasifikasi tabular.

Kita akan melakukan pencarian kombinasi hyperparameter terbaik untuk model **Random Forest** menggunakan GridSearchCV. Tujuannya adalah memaksimalkan performa model pada data yang tidak terlihat (test set).

### Melakukan Grid Search untuk mencari hyperparameter terbaik

```python
# Definisikan parameter grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Inisialisasi GridSearch
grid_search = GridSearchCV(
    estimator=RandomForestClassifier(random_state=42),
    param_grid=param_grid,
    cv=5,
    n_jobs=-1,
    scoring='f1',  # karena kita concern terhadap dropout
    verbose=1
)

# Jalankan pencarian
grid_search.fit(X_train, y_train)

# Model terbaik
best_model = grid_search.best_estimator_

print("Best parameters:", grid_search.best_params_)
```

Melakukan Training Model menggunakan Parameter yang didapatkan dari GridSearch

```python
# Bangun model dengan parameter terbaik
best_model = RandomForestClassifier(
    n_estimators=300,
    max_depth=None,
    min_samples_split=2,
    min_samples_leaf=1,
    random_state=42
)

# Latih ulang model
best_model.fit(X_train, y_train)
```

Simpan Model dan Scaler untuk digunakan di Streamlit nanti

```python
# Simpan model
joblib.dump(best_model, 'best_model.pkl')

# Simpan scaler juga untuk dipakai di Streamlit
joblib.dump(scaler, 'scaler.pkl')

print("Model dan scaler berhasil disimpan!")
```

## Evaluation

Model akan dievaluasi menggunakan data testing (20% dari dataset awal). Metrik utama yang digunakan adalah:

- **Accuracy**: Seberapa banyak prediksi yang benar
- **Precision**: Kemampuan model dalam menghindari false positives
- **Recall**: Kemampuan model dalam menangkap semua kasus dropout
- **F1-score**: Harmoni antara precision dan recall


```python
# Prediksi
y_pred = best_model.predict(X_test)

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Graduate", "Dropout"])
disp.plot(cmap='Blues')
plt.title("Confusion Matrix")
plt.grid(False)
plt.show()
```

```python
# Classification report
print("Classification Report:\n")
print(classification_report(y_test, y_pred, target_names=["Graduate", "Dropout"]))
```
